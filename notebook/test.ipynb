```python
# import data and cosine similarity matrix
import joblib
data=joblib.load('..\\model and data\\data.pkl')
cosine_sim = joblib.load('..\\model and data\\cosine_sim.pkl')
```

```python
# define function to get recommendations
def get_recommendations(title, cosine_sim=cosine_sim):
    title = title.lower()
    if title not in data['title'].values:
        return "Movie not found"
    
    idx = data[data['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return data['title'].iloc[movie_indices]
```

```python
# get user input and print recommendations
print(get_recommendations(input("Enter movie name: ")))
```
